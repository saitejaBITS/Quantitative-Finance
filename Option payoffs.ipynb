```python
import numpy as np
import matplotlib.pyplot as plt
#To make appealing graphs
plt.style.use('seaborn-darkgrid')
```

```python
def put_payoff_buyer(spot, sT , premium):
    p = np.where(spot < sT , sT - spot , 0)
    return p - premium
```

```python
#generating random spot price and fixing strike price , premium

#Microsoft spot price
spot_price = 500

#Put option strike price and premium
strike_price = 500
premium = 15

#let's say that the spot price at the expiration lies in +/-10% from the given spot price
spot = np.arange(0.9*spot_price,1.1*spot_price)
```

```python
#plotting long_put payoff
payoff_longput = put_payoff_buyer(spot,strike_price,premium)

fig,ax = plt.subplots(figsize=(10,6))
ax.spines['bottom'].set_position('zero')
#ax.spines['left'].set_position('zero')
#ax.spines['right'].set_visible(False)
ax.plot(spot,payoff_longput,label = 'long put buyer payoff')
#ax.xaxis.set_ticks_position('bottom')
#ax.yaxis.set_ticks_position('left')
plt.xlabel('Microsoft stock price')
plt.ylabel('profit ~ loss')
plt.legend()
plt.show()
```

```python
payoff_shortput = payoff_longput * -1.0
#Plotting it
fig,ax = plt.subplots(figsize = (10,6))
ax.spines['bottom'].set_position('zero')
ax.plot(spot , payoff_shortput , label = 'short put buyer payoff',color = 'r')
ax.xlabel = 'Microsoft'
ax.ylabel = 'profit ~ loss'
plt.legend()
plt.show()
```

```python
#call option pay off
def call_payoff_buyer(spot , strike_price , premium):
    p1 = np.where(spot > strike_price , spot - strike_price, 0)
    return p1 - premium
```

```python
#plotting long call
payoff_longcall = call_payoff_buyer(spot , strike_price,premium)
fig,ax = plt.subplots(figsize = (10,6))
ax.spines['bottom'].set_position('zero')
ax.plot(spot , payoff_longcall , label = 'long call buyer payoff')
ax.xlabel = 'Microsoft'
ax.ylabel = 'profit ~ loss'
plt.legend()
plt.show()
```

```python
#plotting short call
payoff_shortcall = payoff_longcall * -1.0
fig,ax = plt.subplots(figsize = (10,6))
ax.spines['bottom'].set_position('zero')
ax.plot(spot , payoff_shortcall , label = 'short call buyer payoff' , color = 'r')
ax.xlabel = 'Microsoft'
ax.ylabel = 'profit ~ loss'
plt.legend()
plt.show()
```
